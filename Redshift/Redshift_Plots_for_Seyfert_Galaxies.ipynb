```python
def LaTexCode(dataset):
    roundfactor=2
    stri=""
    for factor in dataset:
        stri=stri+"&"+str(round(factor,roundfactor))
    stri=stri+"&"+str(round(dataset["max"]-dataset["min"],roundfactor))
    return stri.strip("&")
```
