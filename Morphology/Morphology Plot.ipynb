```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convert(list):
    return tuple(list)
```

```python
arr1 = []
arr2 = []
df = pd.read_csv('MorphFile.csv')
df = df.drop(['oid','morph_qual'], axis = 1)
df = df.dropna()
arr_to_check = ['E0', 'E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7', 'S0', 'Sa', 'Sb', 'Sc', 'SBa', 'SBb', 'SBc', 'Irr']
#morph= []
#for i, row in df.iterrows():
    #if ((row['morph_type']) in arr_to_check):
        #morph.append(row['morph_type'])
new_Sy1=df[df['otype']=="Sy1" ]
new_Sy2=df[df['otype']=="Sy2"]
for i in arr_to_check:
    new_Sy1 = df[df['otype']=="Sy1" ]
    new_Sy2 = df[df['otype']=="Sy2"]
    new_Sy1 = new_Sy1[new_Sy1['morph_type'] == i]
    if not new_Sy1.empty:
        new_df = new_Sy1['morph_type']
        for i in range(int(new_df.count())):
            arr1.append(new_df.iloc[i])
    new_Sy2 = new_Sy2[new_Sy2['morph_type'] == i]
    if not new_Sy2.empty:
        new_new_df = new_Sy2['morph_type']
        for i in range(int(new_new_df.count())):
            arr2.append(new_new_df.iloc[i])

#print(arr2)

#print(df.value_counts(['otype', 'morph']))
#bottom label for each bar
#morph = convert(morph)




```

```python
check_Sy1_dict = {
    'E0': 0,
    'E1': 0,
    'E2': 0,
    'E3': 0,
    'E4': 0,
    'E5': 0,
    'E6': 0,
    'E7': 0,
    'S0': 0,
    'Sa': 0,
    'Sb': 0,
    'Sc': 0,
    'SBa': 0,
    'SBb': 0,
    'SBc': 0,
    'Irr': 0
}

for i in arr1:
    check_Sy1_dict[i] += 1

check_Sy2_dict = {
    'E0': 0,
    'E1': 0,
    'E2': 0,
    'E3': 0,
    'E4': 0,
    'E5': 0,
    'E6': 0,
    'E7': 0,
    'S0': 0,
    'Sa': 0,
    'Sb': 0,
    'Sc': 0,
    'SBa': 0,
    'SBb': 0,
    'SBc': 0,
    'Irr': 0
}

for j in arr2:
    check_Sy2_dict[j] += 1
```

```python
x =  ['E0', 'E1','E2','E3', 'E4', 'E5', 'E6', 'E7', 'S0', 'Sa', 'Sb', 'Sc', 'SBa', 'SBb', 'SBc', 'Irr']
Sey1 = []
Sey2 = []

#print(check_Sy1_dict)
#print(check_Sy2_dict)
for i in x:
    Sey1.append(check_Sy1_dict[i])
    Sey2.append(check_Sy2_dict[i])

y1 = np.array(Sey1)
y2 = np.array(Sey2)

plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')

plt.xlabel("Galaxy Types")
plt.ylabel("Number of Galaxies")
plt.legend(["Seyfert 1", "Seyfert 2"])
plt.title("Seyfert Galaxies by Morphology")
plt.show()
```
