```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv("LuminositySeyfertData.csv")
print(df)
```

```python
df = df.dropna(subset=['V'])
#df = df.dropna(subset=['rvz_redshift'])
new_Sy1=df[df['otype']=="Sy1"]
new_Sy2=df[df['otype']=="Sy2"]

s1 = new_Sy1.V.to_numpy()
s2 = new_Sy2.V.to_numpy()

n_bins = 10
_, bins, _ = plt.hist(s1, bins=n_bins, edgecolor='black')

# Use the same bin edges for the other dataset
plt.hist(s1, bins=bins, edgecolor='black', alpha=0.5, label='Sy1')
plt.hist(s2, bins=bins, edgecolor='black', alpha=0.5, label='Sy2')

plt.xlabel('V Magnitude')
plt.ylabel('Frequency')
plt.title('Histogram of V Magnitude')

plt.show()
```
